```python
import pandas as pd
df=pd.read_excel('반도체_공정설계_4차과제_raw_data.xlsx',skiprows=2,usecols='B:N')
df
```

```python
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size']=15
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
print(df.columns.tolist())
```

```python

df.columns = df.columns.str.strip()

# 결측값 제거
df_clean = df.dropna(subset=['Equipment ID', 'Value(Å) (average)', 'Target'])

# Target 값 
target_value = df_clean['Target'].iloc[0]

# 박스플롯 그리기
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x='Equipment ID', y='Value(Å) (average)', data=df_clean)

# Target 점선 추가
plt.axhline(y=target_value, color='red', linestyle='--', linewidth=2, label=f'Target ({target_value:.0f} Å)')


plt.title('ILD TEOS Thickness by Equipment ID')
plt.xlabel('Equipment ID')
plt.ylabel('ILD TEOS Thickness (Å)')
plt.ylim(1900, 2100)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()



```
